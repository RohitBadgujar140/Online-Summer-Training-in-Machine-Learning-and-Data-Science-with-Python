# Support Vector Machine Algorithm (Supervised Machine Learning)

It's Supervised Machine Learning Algorithm and used for both regression and mostly used for classification

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
Import Dataset from sklearn
from sklearn.datasets import load_breast_cancer
```

```python
# Load Dataset (It is in Dictionsary form)
cancer = load_breast_cancer()
```

```python
# return the name of the columns in the dataset (this is n-dimension data) Features(x)
cancer.feature_names
```

```python
# Return labels name(y) 
cancer.target_names
```

```python
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
df.head()
```

```python
df['target'] = cancer.target
df.tail()
```

```python
df[df['target']==1].head()
```

```python
df['cancer class'] = df.target.apply(lambda x: cancer.target_names[x])
df.head()
```

```python
df[50:55]
```

```python
df0 = df[df.target==0].head(50) # class 1 data
df1 = df[df.target==1].head(50) # class 2 data
```

```python
import matplotlib.pyplot as plt
plt.xlabel('mean radius')
plt.ylabel('mean texture')
plt.scatter(df0['mean radius'], df1['mean texture'], color='red', marker='+')
plt.scatter(df1['mean radius'], df0['mean texture'], color='green', marker='*')
```

```python
# Divide dataset into x(features as i/p) and y(label as o/p)
x = df.drop(['target', 'cancer class'], axis='columns') # x is features 2D Numpy array
y = df.target                                            # y is lablel   1D Numpy array
```

```python
# Divide dataset(x,y) into testing data as xtest,ytest and training data as xtrain,ytrain
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
# create model support vector machine
from sklearn.svm import SVC
model = SVC()
```

```python
# Train our model by providing training data
Mymodel = model.fit(xtrain,ytrain)
Mymodel
```

```python
# Accuracy of model
Mymodel.score(xtest,ytest)
```

```python
# Prediction by our model
y_pred = Mymodel.predict(xtest)
```

```python
# Difference of y_pred(predicted data by our model) and ytest(actual data in dataset)
df = pd.DataFrame(y_pred, ytest)
df.head()
```

```python
#Importing confusion matrix

from sklearn.metrics import confusion_matrix

cm=np.array(confusion_matrix(ytest,y_pred))

cm
```

```python
# Regularization aim to fit training set data and avoid overfitting
model_c = SVC(C=1)
model_c.fit(xtrain,ytrain)
model_c.score(xtest,ytest) # accuracy
```

```python
model_c = SVC(C=2)
model_c.fit(xtrain,ytrain)
model_c.score(xtest,ytest) # accuracy
```

```python
# Kernel function does is convert data from low dimensions to high dimensions.
model_kernel = SVC(kernel='linear')
model_kernel.fit(xtrain,ytrain)
model_kernel.score(xtest,ytest)
```
