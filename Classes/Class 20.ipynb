# Web Scraping in Python

Steps

1) Find URL

2) Inspecting the Page

3) Find data you want to extract

4) Write code

5) Execute code

6) Store data

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=806895a4-ea49-4154-ae7b-7d8087fb9141"
```

```python
response = requests.get(url)
print(response)
# print(response.content) print html content
htmlcontent = response.content
```

```python
soup = BeautifulSoup(htmlcontent, 'html.parser')
print(soup.prettify)
```

```python
products = []
prices = []
reviews = []

#product = soup.find('div', attrs={'class':'_4rR01T'})
#print(product.text)
```

```python
for i in soup.find_all('a', href=True, attrs={'class':'_1fQZEK'}):
    try:
        product = i.find('div', attrs={'class':'_4rR01T'})
        price = i.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating = i.find('div', attrs={'class':'_3LWZlK'})
        products.append(product.text)
        prices.append(price.text)
        reviews.append(rating.text)
    except:
        reviews.append('None')
        print('Exception')

print(set(products))
print(set(prices))
print(set(reviews))
```

# Store Result in Particular Format

```python
df = pd.DataFrame({'Product Name':products, 'Prices':prices, 'Reviews':reviews})
df.head()
```

```python
df.to_csv('laptops.csv')
```
