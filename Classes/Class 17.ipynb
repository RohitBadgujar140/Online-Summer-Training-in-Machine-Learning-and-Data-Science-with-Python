# Exploratory Data Analysis on Haberman Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("haberman.csv")
df.head()
```

```python
df.info()
```

```python
df['status'].value_counts() # To check balanced or imbalanced data using value_counts() on last column
```

```python
df.isnull()
```

```python
sns.heatmap(df.isnull(), yticklabels=False) # check null values with the help of heatmap
```

```python
df.isnull().sum()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='status', data=df)
```

```python
df.describe()
```

## 1) Bivariate Analysis

```python
sns.set_style('whitegrid')
sns.FacetGrid(df, hue='status').map(plt.scatter,'age','nodes').add_legend()
```

```python
# for multible variables plot pair plot (It's draw only for numerical columns)
sns.pairplot(df, hue='status', vars=['age','year','nodes'], palette='gist_rainbow')
```

## 2) Univariate Analysis

```python
live = df.loc[df['status']==1]
death = df.loc[df['status']==2]

plt.plot(live['age'],np.zeros_like(live['age']), 'o')
plt.plot(death['age'],np.zeros_like(death['age']), 'o')
```

```python
sns.FacetGrid(df, hue='status').map(sns.distplot,'year').add_legend()
```

```python
sns.distplot(df['age'], color='darkred')
```
