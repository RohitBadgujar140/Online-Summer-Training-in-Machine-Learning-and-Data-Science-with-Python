5) Data Selection – Based on Conditional filtering

6) Groupby operations

```python
df = imdb.groupby('Genre').mean()
df
```

```python
imdb[imdb['Rating']>7].groupby('Genre')[['Rating']].count()
```

```python
top_movie = imdb[imdb["Rating"] > 8.0]
top_movie.groupby(["Title"])["Votes"].mean()
```

```python
x = imdb.sort_values(by='Rating')

x.head(10)
```
